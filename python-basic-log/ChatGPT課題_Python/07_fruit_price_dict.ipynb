```python
#課題：フルーツ価格表に新しいフルーツを追加して表示する

#フルーツの名前と価格を「辞書（dict）」で管理しなさい。

#その辞書に、新しいフルーツと価格を追加しなさい。

#追加後の辞書を使って、次の処理を行うこと。

#フルーツ名と価格を1行ずつ表示する

#表示形式は「りんご: 120円」のようにする
```

```python
# フルーツ価格表を辞書で管理
fruits = {
    "りんご": 120,
    "みかん": 80,
    "バナナ": 100,
    "ぶどう": 350
}

print("🍎 フルーツ価格表（追加前） 🍇")
for fruit, price in fruits.items():
    print(f"{fruit}: {price}円")

# 新しいフルーツを追加
fruits["メロン"] = 500
fruits["いちご"] = 400

print("\n🍎 フルーツ価格表（追加後） 🍇")
for fruit, price in fruits.items():
    print(f"{fruit}: {price}円")

# "\n" を入れると 1行分の空白行 が入ってから次の文字列が表示されます。👇
```

```python
# 辞書（dict） は「キー: 値」の組み合わせで管理できる

# ここでは「フルーツ名」がキー、「価格」が値

# fruits.items() を使うと、キーと値を同時に取り出せる

# f文字列（f"{...}"）を使うと、文字列の中に変数を埋め込めて便利
```
