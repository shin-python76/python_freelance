```python
#課題
#1から100までの整数の合計を求めて、出力しなさい。

#for文を使った方法

#while文を使った方法
```

```python
#for文を使った方法

total = 0                 # ① 合計の入れ物を0で初期化（累積変数）

for i in range(1, 101):   # ② 1から100までの整数を順にiに代入しながら繰り返す
    total += i            # ③ total = total + i（累積：iを加える）
                          # ④ iは自動的に1ずつ増えていき、101になった時点でループ終了

print('1から100までの合計は', total)  # ⑤ ループが終わったら結果を表示

```

```python
#while文を使った方法

total = 0     # ① 合計の入れ物を0で初期化（累積変数）
i = 1         # ② いま足そうとしている数（カウンタ）を1から開始

while i <= 100:        # ③ ループ継続条件：「iが100以下の間は繰り返す」
    total += i         # ④ total = total + i（累積：iを加える）
    i += 1             # ⑤ カウンタを1進める（次はi+1を足す）
                       # ⑥ ③に戻って条件判定 → Trueなら④⑤をまた実行

print('1から100までの合計は', total)  # ⑦ ループが終わったら結果を表示
```

```python
# forとの違い（メンタルモデル）

# for i in range(1, 101)は**「1から100まで順にiをくれる装置」**。

# while i <= 100は**「自分でiを管理して、条件が満たされる限り手動で回す」。
# だからwhileは初期化（i=1）と更新（i+=1）**を自分で書く必要があります。
```
