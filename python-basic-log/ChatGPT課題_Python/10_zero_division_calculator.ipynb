```python
# ================================================================
# 🧮 課題：【0除算エラーを防ぐ電卓】
# ================================================================

# 【1】目的
# ------------------------------------------------
# ・try〜except構文の基本的な使い方を学ぶ
# ・0で割り算した場合のエラー（ZeroDivisionError）を防ぐ
# ・プログラムが強制終了せず、ユーザーに分かりやすく通知できるようにする


# 【2】関数の定義
# ------------------------------------------------
# 2つの数を受け取り、割り算を行う関数を定義する
# "try" で通常の処理を行い、"except" でエラー発生時の処理を書く

def divide(a, b):
    try:
        result = a / b
        print(f'計算結果: {result}')
    except ZeroDivisionError:
        print('⚠️ エラー: 0では割り算できません！')


# 【3】動作確認
# ------------------------------------------------
# 正常なケースと、エラーを起こすケースをそれぞれ試してみる

divide(10, 2)   # 正常：10 ÷ 2 = 5.0
divide(5, 0)    # エラー：0で割り算 → exceptが実行される


# 【4】実行結果（例）
# ------------------------------------------------
# 計算結果: 5.0
# ⚠️ エラー: 0では割り算できません！


# 【5】学習メモ
# ------------------------------------------------
# ・tryブロック：通常の処理を実行
# ・exceptブロック：特定のエラーをキャッチして代替処理を実行
# ・ZeroDivisionError：0で割ったときに発生するエラー
# ・エラーを個別に指定することで、プログラムの信頼性が向上する
```
