```python
#課題：成績リストから平均を出す

#いくつかのテストの点数をリストに入れなさい。

#そのリストを使って、次の処理を行うこと。

#合計点を計算して表示する

#平均点を計算して表示する（小数点1桁まで表示）
```

```python
scores = [50, 40, 70, 60, 80]
scores
```

```python
print(sum(scores))
```

```python
# 合計点
total = sum(scores)
print("合計:", total)

# 平均点（小数第1位まで表示）
average = total / len(scores)

# リストの合計点数をリストの要素の数で割る

print("平均:", round(average, 1))

# round関数とは
# round(数値, 桁数)
# 数値 … 丸めたい数
# 桁数 … 小数点以下を何桁残すか
```
