```python
# =============================================================================
# 🎮 課題：[簡易じゃんけんゲーム]
# =============================================================================

# 【1】目的
# -----------------------------------------------------------------------------
# ・条件分岐（if / elif / else）の練習をする
# ・input()でユーザー入力を受け取り、randomモジュールでCPUの手を決定する
# ・勝敗を判定し、結果をprint()で表示する
# ・プログラムの流れを「入力 → 処理 → 出力」として体験する


# 【2】モジュールの準備
# -----------------------------------------------------------------------------
import random  # CPUの手をランダムに決めるために使用

# 👆random は、乱数処理のための“モジュール”


# 【3】関数の定義
# -----------------------------------------------------------------------------
# ・ユーザーの手とCPUの手を受け取り、勝敗を判定する関数を作成
# ・returnで結果（文字列）を返す

def judge(player, cpu):
    if player == cpu:
        return "あいこ"
    elif (player == "グー" and cpu == "チョキ") or \
         (player == "チョキ" and cpu == "パー") or \
         (player == "パー" and cpu == "グー"):
        return "あなたの勝ち！"
    else:
        return "あなたの負け..."


# 【4】メイン処理
# -----------------------------------------------------------------------------
# ・ユーザーから手を入力してもらい、CPUの手をランダムに決定する
# ・judge関数を使って結果を表示する

hands = ["グー", "チョキ", "パー"]

# じゃんけんの選択肢を、hands というリスト[]にまとめる

print("=== じゃんけんゲーム ===")
player = input("あなたの手を入力してください（グー／チョキ／パー）：")

# ユーザーが入力した文字列（”グー”や”チョキ”など）が　player に代入される

# 入力チェック
if player not in hands:
    print("⚠️ 入力が正しくありません。『グー』『チョキ』『パー』のいずれかを入力してください。")

# 入力が正しいかどうかをチェックしています。
# in は「含まれているか？」を調べる演算子。
# not in は「含まれていないか？」を調べます。
#つまり：
# if player not in ["グー", "チョキ", "パー"]:
# ➡️ これで、「グー・チョキ・パー以外なら警告を出す」という条件になります。

else:
    cpu = random.choice(hands)
    print(f"コンピュータの手：{cpu}")
    result = judge(player, cpu)
    print(f"結果：{result}")


# random.choice(リスト) は、リストの中からランダムに1つ選びます。
# たとえば "チョキ" が選ばれた場合、cpu = "チョキ" となります。

# print(f"コンピュータの手：{cpu}")
# ➡️ f文字列を使って、選ばれた手を画面に表示しています。
# {cpu} の部分に、実際の値（例：「チョキ」）が埋め込まれます。

# result = judge(player, cpu)
# ➡️ ここで先ほど定義した関数 judge() を呼び出します。
# player（あなたの手）と cpu（コンピュータの手）を渡し、
# judge() が判定した結果（例：「あなたの勝ち！」）を返します。
# その返り値を result に代入します。


# 【5】動作確認例
# -----------------------------------------------------------------------------
# 実行例１：
# あなたの手を入力してください（グー／チョキ／パー）：グー
# コンピュータの手：チョキ
# 結果：あなたの勝ち！
#
# 実行例２：
# あなたの手を入力してください（グー／チョキ／パー）：パー
# コンピュータの手：パー
# 結果：あいこ
```
