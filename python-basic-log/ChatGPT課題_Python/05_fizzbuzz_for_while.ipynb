```python
#課題

#1から100までの整数を順に出力するプログラムを作成しなさい。
#ただし、次の条件に従うこと。

#3の倍数のときは「Fizz」と出力する

#5の倍数のときは「Buzz」と出力する

#15の倍数のときは「FizzBuzz」と出力する
```

```python
# for文を使った方法

for i in range(1, 101):       # ① 1から100までを順にiに代入
    if i % 15 == 0:           # ② 15の倍数（3と5の両方で割り切れる）
        print("FizzBuzz")     #    まず「15の倍数」を先に書く（15がFizzやBuzzで止まってしまう）。
    elif i % 3 == 0:          # ③ 3の倍数
        print("Fizz")
    elif i % 5 == 0:          # ④ 5の倍数
        print("Buzz")
    else:                     # ⑤ それ以外の数
        print(i)
```

```python
# while文を使った方法

i = 1                      # ① カウンタを1からスタート

while i <= 100:            # ② iが100以下の間、繰り返す
    if i % 15 == 0:         # ③ 15の倍数（3と5の両方で割り切れる）
        print("FizzBuzz")
    elif i % 3 == 0:        # ④ 3の倍数
        print("Fizz")
    elif i % 5 == 0:        # ⑤ 5の倍数
        print("Buzz")
    else:                   # ⑥ それ以外の数
        print(i)

    i += 1                  # ⑦ カウンタを1進める
```

```python
# while文版のポイント

# 初期化（i = 1） と 更新（i += 1） を必ず自分で書く必要がある。

# 忘れると無限ループになります⚠️

# 条件の順番はfor文版と同じで、15の倍数を先に判定するのが重要です。
```
