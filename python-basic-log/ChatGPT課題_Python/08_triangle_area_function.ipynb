```python
# 課題：三角形の面積を計算する関数

# 条件

# 三角形の面積を計算する関数を作成しなさい。

# 関数名は自由とするが、引数として「底辺（base）」と「高さ（height）」を受け取ること。

# 面積の計算は「底辺 × 高さ ÷ 2」の公式を使うこと。

# 関数は計算した面積を return で返すこと。
```

```python
# 【三角形の面積を計算する関数を定義】

# triangle_area() を呼び出したとき、area が結果として返る
# 👉 もし return がなければ、関数の中で計算しても結果を外に出せない
#（printは表示するだけ、returnは“値を外に渡す”もの、と覚えるとわかりやすい）

def triangle_area(base, height):
  area = base * height / 2
  return area

# 【ユーザーから底辺と高さを入力してもらう】
# input("...")
# 画面にプロンプト（案内文）を表示して、ユーザーの入力が終わるまで待機
# ユーザーが Enter を押すと、**文字列（str）**として返る。改行は返り値に含まれない。

# float(返ってきた文字列)
# 文字列を**浮動小数点数（float）**に変換する。
# 例："10"→10.0、"5.5"→5.5、"1e3"→1000.0。
# 先頭末尾の空白は無視される（" 12.5 " もOK）。

# base = ... / height = ...
# 変換できた数値が、それぞれの変数に代入される。以後は数値として計算に使える。


base = float(input('底辺の長さを入力してください：'))
height = float(input('高さを入力してください：'))


# 【関数を呼び出して面積を計算】

# 関数を呼び出し、その関数が返した「戻り値」を result という変数に代入。
# これで、関数の結果をプログラムの外側で自由に使えるようになる。

result = triangle_area(base, height)

# 結果を表示

print('三角形の面積は', result, 'です。')


# ⭐️なぜ float() が必要？

# input() の戻りは必ず文字列です。
# 文字列のままでは数式に使えず、例えば base * height / 2 をするとエラーになります（TypeError）。
# float() で数値化しておくと、小数も扱えて面積計算に最適です。
# ※整数だけで良いなら int() でもOKですが、小数入力（例：7.5）に対応できません。
```
